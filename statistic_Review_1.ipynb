## Measures of Central Tendency

### mean
The mean is calculated by summing all values and dividing by the total number of values. It's useful for evenly distributed data.

Example:
Imagine splitting a $100 meal among 5 friends. To find how much each person pays:

(20+25+15+30+10)÷5=100÷5=20

Each person pays $20.



```python
mean = sum(num_friends) / len(num_friends)
print(mean)
```

###median

The median is the middle number when data is ordered. If there’s an even number of values, it’s the average of the two middle numbers.

Example:
For 5 box weights (3kg, 5kg, 8kg, 10kg, 12kg),

the median is 8kg.

For 6 box weights (3kg, 5kg, 8kg, 10kg, 12kg, 15kg),

the median is:
(8+10)÷2=9

```python
# def of function called median from principle
def median(values):
    n = len(values)
    sorted_values = sorted(values)  # ascending order NECESSARY
    midpoint = n // 2

    if n % 2 == 1:
        return sorted_values[midpoint]

    else:
        lower = midpoint - 1
        upper = midpoint
        return (sorted_values[lower] + sorted_values[upper]) / 2
```

```python
median(num_friends)
```

### quantil (the generalization of median)
A quantile is a value that divides a data set into equal parts, based on a given probability. For example, the median is a quantile that divides the data into two equal halves, while other quantiles, like quartiles or percentiles, divide the data into four or one hundred equal parts, respectively. The specific quantile corresponds to a percentage or probability of data points falling below that value.


```python
# Quantile Definition
def quantile(x, p):
    p_index = int(p * len(x))  # Calculate the index for the quantile by multiplying 'p' with the length of the list
    return sorted(x)[p_index]  # Return the element at the calculated index from the sorted list 'x'
```

### Mode

The mode is the value that appears most often. Datasets can have one mode, multiple modes, or none at all.

Example:
If pet counts in 10 homes are: 1, 2, 2, 3, 1, 4, 2, 1, 3, 2, the mode is 2, as it appears 4 times.

```python
#Mode definition

def mode(x):
    counts = Counter(x)     # Creates a Counter object that counts the occurrences of each value in 'x'
    max_count = max(counts.values())     # Finds the maximum frequency of occurrences
    return [x_i for x_i, count in counts.items()
    if count == max_count]
```

```python
print(mode(num_friends))
```

## Dispersion

### amplitude
The range measures the spread between the smallest and largest values in a dataset.

Range=Maximum Value - Minimum Value

Example:
For the dataset {3, 7, 10, 15}, the range is:

15-3=12

It's simple but sensitive to outliers.

```python
# Amplitude Definition
def data_range(x):
  return max(x) - min(x)
```

```python
print(data_range(num_friends))
```

### variance
Variance measures how far data points are from the mean, on average. It's the average of the squared differences from the mean.

Example:
For {2, 4, 6}:

Mean:
(2+4+6)÷3 = 4

Differences:
(2-4)²,(4-4)²,(6-4)²=4,0,4

Variance:
(4+0+4)÷3=2.67


```python
def dot(v, w):
  return sum(v_i * w_i
             for v_i, w_i in zip(v, w))
```

```python
def sum_of_squares(x):
  return dot(x,x)
```

```python
def de_mean(x):
  x_bar = sum(x) / len(x) # Calculate the mean directly within the function
  return [x_i - x_bar for x_i in x]
```

```python
# Variance Definition
def variance(x):
  n = len(x)
  deviations = de_mean(x)
  return sum_of_squares(deviations) / (n - 1)
```

```python
print(variance(num_friends))
```

### Standard deviation
The standard deviation is the square root of variance. It's easier to interpret because it's in the same units as the data (IMPORTANT).



```python
# STD definiton
def standard_deviation(x):
  return math.sqrt(variance(x))
```

```python
standard_deviation(num_friends)
```

### Interquartile

The interquartile range (IQR) measures the spread of the middle 50% of your data. It is calculated as:

IQR=Q3-Q1
Where:

Q1 (First Quartile): The value below which 25% of the data falls.

Q3 (Third Quartile): The value below which 75% of the data falls.

The IQR is helpful because it ignores outliers (extremely high or low values), focusing only on the central portion of the dataset.

Interpretation:
The IQR tells us that the middle 50% of the data values lie within a range of 8 units (from 3 to 11). This makes it easier to understand the central spread of the data without being affected by potential outliers like 1 or 13.

Why Use the IQR?
The IQR is especially useful when your data has outliers, as it focuses on the range where most of your data lies, ignoring extreme values. For instance:

In salaries, an outlier (e.g., a CEO's income) won’t affect the IQR.
In test scores, it helps identify the consistent performance of most students.



```python
# Interquartile Definition
def interquartile_range(x):
  return quantile(x, 0.75) - quantile(x, 0.25)
```

```python
print(interquartile_range(num_friends))
```

## Correlation
Correlation measures the strength and direction of the relationship between two variables. Its value ranges from -1 to 1.

+1: Perfect positive correlation.

-1: Perfect negative correlation.

0: No correlation.

Example:
Height and shoe size have a positive correlation, while hours of exercise and body weight may have a negative correlation.



```python
# Covariance Definition
def covariance(x, y):
  n = len(x)
  return dot(de_mean(x), de_mean(y))/(n-1)
```

```python
daily_minutes = []
for num in num_friends:
  daily_minutes.append(num * 10) # Assuming 10 minutes per friend

daily_minutes
```

```python
daily_minutes.sort() # Sort daily_minutes in ascending order
daily_minutes
```

```python
print(covariance(num_friends, daily_minutes))
```

```python
# Correlation Definition
def correlation(x,y):
  stdev_x = standard_deviation(x)
  stdev_y = standard_deviation(y)

  if stdev_x > 0 and stdev_y > 0:
    return covariance(x,y)/stdev_x/stdev_y
  else:
    return 0
```

```python
correlation(num_friends, daily_minutes)
```

## Simpson Paradox

Simpson's paradox is a statistical phenomenon where a trend observed in individual groups of data reverses or disappears when these groups are combined. This occurs due to the influence of hidden variables or confounding factors, which can distort the interpretation of the data.
